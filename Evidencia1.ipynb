```python
#Identificar valores nulos
null=datos.isnull().sum()
null
```

```python
#Se elimina la columna de municipio_hechos porque son demasiados valores atípicos
datos = datos.drop(columns=["idCarpeta","municipio_hechos","Año_hecho","Mes_hecho","Año_inicio","Mes_inicio"])
```

```python
datos.columns.values
```

```python
nombres=[
    "fecha_inicio",
    "delito",
    "categoria",
    "sexo",
    "edad",
    "tipo_persona",
    "calidad_juridica",
    "competencia",
    "fecha_hecho",
    "hora_hecho",
    "hora_inicio",
    "alcaldia_hechos",
    "colonia",
    "latitud",
    "longitud"
]
datos.columns=nombres
```

```python
datos.head(5)
```

```python
datos["edad"].describe()
```

```python
#Se reemplazan por la mediana para mantener la distribución de los datos
datos["edad"]=datos["edad"].fillna(round(datos["edad"].median(),0))
```

```python
datos.sexo.value_counts()
```

```python
#Se rellena por el valor anterior de la misma columna, se vuelve a correr la cuenta de valores para verificar que el orden no haya cambiado
datos["sexo"]= datos["sexo"].fillna(method="bfill") 
```

```python
datos.sexo.value_counts()
```

```python
datos.tipo_persona.value_counts()
```

```python
#Se rellena por el siguiente valor de la misma columna
datos["tipo_persona"]= datos["tipo_persona"].fillna(method="ffill") 
```

```python
datos.tipo_persona.value_counts()
```

```python
datos.calidad_juridica.value_counts()
```

```python
#Se rellena por el siguiente valor como string de la calidad jurídica más común para no alterar los resultados
datos["calidad_juridica"]= datos["calidad_juridica"].fillna("VICTIMA Y DENUNCIANTE") 
```

```python
datos.fecha_hecho.value_counts()
```

```python
#Se rellena por el siguiente valor como string de la fecha más común para no alterar los resultados
datos["fecha_hecho"]= datos["fecha_hecho"].fillna("2021-01-01") 
```

```python
datos.hora_hecho.value_counts()
```

```python
#Se rellena por el siguiente valor como string de la hora más común para no alterar los resultados
datos["hora_hecho"]= datos["hora_hecho"].fillna("12:00:00") 
```

```python
datos.hora_inicio.value_counts()
```

```python
#Se rellena por el siguiente valor como string de la hora inicio más común para no alterar los resultados
datos["hora_inicio"]= datos["hora_inicio"].fillna("14:27:00") 
```

```python
datos.colonia.value_counts()
```

```python
#Se rellena por el siguiente valor como string de la colonia más común para no alterar los resultados
datos["colonia"]= datos["colonia"].fillna("CENTRO VIII") 
```

```python
datos.competencia.value_counts()
```

```python
datos["competencia"]=datos["competencia"].replace("HECHOS NO DELICTIVOS","HECHO NO DELICTIVO")
```

```python
#Identificar valores nulos
null=datos.isnull().sum()
null
```
